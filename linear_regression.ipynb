```python
#import the necessory libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df
```

```python
%matplotlib inline

plt.xlabel("year")
plt.ylabel("per capita income")

plt.scatter(df.year,
            df['per capita income (US$)'],
            color='purple',
            marker='*')
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[['year']],
        df['per capita income (US$)'])
```

```python
reg.predict(pd.DataFrame({'year':[2020]}))
```
