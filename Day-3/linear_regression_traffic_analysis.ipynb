**Linear Regression**

**Problem Statement**

This project uses Single Linear Regression to predict traffic risk score based on the number of vehicles involved. The system analyses traffic data and predicts how accident risk increases when vehicle count increases.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("traffic_single_linear_dataset.csv")

df
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
%matplotlib inline

plt.xlabel('Vehicles Involved')
plt.ylabel('Risk Score')

plt.scatter(df.vehicles_involved,
            df.risk_score,
            color='green',
            marker='*')
```

```python
new_df = df.drop('risk_score',
                 axis='columns')

new_df
```

```python
risk = df.risk_score

risk
```

```python
reg = LinearRegression()
```

```python
reg.fit(new_df, risk)
```

```python
vehicles_involved = 10
```

```python
reg.predict([[10]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
10 * reg.coef_ + reg.intercept_
```

```python
plt.xlabel('Vehicles Involved')
plt.ylabel('Risk Score')

plt.scatter(df.vehicles_involved,
            df.risk_score,
            color='green',
            marker='*')

plt.plot(df.vehicles_involved,
         reg.predict(new_df),
         color='purple')
```
