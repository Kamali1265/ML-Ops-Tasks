**Multiple Linear Regression**

**Problem Statement**

Predict traffic density based on multiple traffic-related factors such as vehicle count, average speed, and horn count using Multiple Linear Regression.

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv("traffic_multiple_lr.csv")

df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[['Vehicle_Count',
            'Avg_Speed',
            'Horn_Count']],

        df['Traffic_Density'])
```

```python
reg.predict([[120, 40, 35]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.coef_[0]*120 + \
reg.coef_[1]*40 + \
reg.coef_[2]*35 + \
reg.intercept_
```

```python
reg.predict([[180, 30, 50]])
```
